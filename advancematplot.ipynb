Subplotes means more than one plot in a graph

```python
fig ,ax = plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0].set_title('Avg Vs Strike_rate')
ax[0].set_ylabel('Strike_Rate')

ax[1].scatter(batters['avg'],batters['runs'],color='green')
ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

We can make four or more graphs at a time

```python
fig ,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,0].set_title('Avg vs Strike')
ax[0,1].scatter(batters['avg'],batters['runs'],color='green',marker='+')
ax[0,1].set_title('Avg vs Runs')
ax[1,0].hist(batters['avg'],color='blue')
ax[1,0].set_title('Avg')
ax[1,1].hist(batters['runs'])
ax[1,1].set_title('Runs')

```

```python
x = [0,23,12,8]
y = [0,24,4,12]
z = [5,0,10,15]

fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')
```

3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

Contour Plot

when you want to convert 3D plot into 2D so that you can understand more

```python
fig = plt.figure(figsize=(8,6))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

Fill contour plot by calling a function contourf

```python
fig = plt.figure(figsize=(8,6))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
fig.show()
```

HeatMap

```python
delivery = pd.read_csv('E:/courses/Data science/Data science journey/matplotlibe/advancematplot/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head(3)
```

Here in temp_df we first take ballnumber which is 1,2,3,4,5,6 in a over

then we take the batsman runs which should be equal to 6

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

Now for displaying in a table we call a function name as pivot_table

we pass "overs" as index "ballnumber" as coloums and we want the values of "batsman_run

we apply the aggregate function to count all the number of sixes per ball

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

Now we have to make a heatmap chart of the varible name as "grid"

```python
plt.figure(figsize=(10,8))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```
